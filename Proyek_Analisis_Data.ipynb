## Visualization & Explanatory Analysis

### Pertanyaan 1:

Di musim apa sewa sepeda paling banyak?

```python
musim_df = jam_df.groupby(by="weathersit").cnt.nunique().reset_index()
musim_df.rename(columns={
    "cnt": "count_sewa"
}, inplace=True)

colors = ["#FFA500", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]
musim_df
plt.figure(figsize=(10,5))
sb.barplot(
    y="count_sewa",
    x="weathersit",
    data=musim_df.sort_values(by="count_sewa", ascending=False),
    palette=colors
)
plt.title("Banyakya Jumlah Sepeda yang disewa pada setiap Musim", loc="center", fontsize=15)
plt.xlabel("Musim")
plt.ylabel("Banyak Sepeda disewa")
plt.show()
```

```python
musim_df = hari_df.groupby(by="weathersit").cnt.nunique().reset_index()
musim_df.rename(columns={
    "cnt": "count_sewa"
}, inplace=True)

colors = ["#FFA500", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]
musim_df
plt.figure(figsize=(10,5))
sb.barplot(
    y="count_sewa",
    x="weathersit",
    data=musim_df.sort_values(by="count_sewa", ascending=False),
    palette=colors
)
plt.title("Banyakya Jumlah Sepeda yang disewa pada setiap Musim", loc="center", fontsize=15)
plt.xlabel("Musim")
plt.ylabel("Banyak Sepeda disewa")
plt.show()
```

### Pertanyaan 2:

Apakah terdapat perbedaan dalam pola penggunaan sepeda antara hari libur dan hari kerja?

```python
# Mengelompokkan data berdasarkan weekday dan holiday, lalu menghitung statistiknya
statistik_penggunaan = jam_df.groupby(['weekday', 'holiday']).agg({
    "registered": ["sum", "max", "min"],
    "casual": ["sum", "max", "min"]
}).reset_index()

print(statistik_penggunaan)
```

```python
# Mengelompokkan data berdasarkan weekday dan holiday, lalu menghitung statistiknya
statistik_penggunaan = hari_df.groupby(['weekday', 'holiday']).agg({
    "registered": ["sum", "max", "min"],
    "casual": ["sum", "max", "min"]
}).reset_index()

print(statistik_penggunaan)
```
